```python
import pandas as pd

data = pd.read_csv("landmarkDataset.csv")
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20)
```

```python
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()
classifier.add(Dense(24,activation="relu",input_dim = 42))
classifier.add(Dense(24,activation="relu"))
classifier.add(Dense(6,activation="softmax"))
classifier.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

```

```python
classifier.fit(x_train,y_train,batch_size=10,epochs=500)
```

```python
import numpy as np
y_pred = classifier.predict(x_test)
y_pred_lables = np.argmax(y_pred,axis=1)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred_lables)
cm
```
